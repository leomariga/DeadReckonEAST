```python
from ITA_DeadReckoning2d import ITA_DeadReckoning2d

estimadorDR = ITA_DeadReckoning2d()
xPred = []
yPred = []
erroalinha = []
aviaoalvo = []
distCargaAlvo = []

# Itera em cada linha
for i, row in df.iterrows():
    x,y = estimadorDR.estimaPosi(row['XGPS'], row['YGPS'], row['HeadGPS'], row['VelocidadeGPS'], row['Tempo'], 4, 64)
    x_carga, y_carga = estimadorDR.estimaPosiCarga(row['ZGPS'], row['HeadGPS'], row['VelocidadeGPS'], 1, 4, 64)
    xPred.append(x)
    yPred.append(y)
    erroalinha.append(estimadorDR.erroAlinhamento)
    aviaoalvo.append(estimadorDR.head_aviao_alvo)
    distCargaAlvo.append(estimadorDR.distCargaAlvo)
    
df.insert(df.shape[1]-1, 'xpreddr', xPred)
df.insert(df.shape[1]-1, 'ypreddr', yPred)
df.insert(df.shape[1]-1, 'erroalinha', erroalinha)
df.insert(df.shape[1]-1, 'distCargaAlvo',distCargaAlvo)
df.insert(df.shape[1]-1, 'aviaoalvo',aviaoalvo)
```

```python
ax = plt.gca()
df.plot(kind='line',x='Tempo',y='PosPayloadDrop_GPS',ax=ax)
df.plot(kind='line',x='Tempo',y='distCargaAlvo',ax=ax)
#ax.set_xlim(65100, 65150)
ax.set_xlim(65100, 65170)
ax.set_ylim(-20, 80)
```
